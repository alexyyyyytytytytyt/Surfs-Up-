# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
from sqlalchemy import extract
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite", echo=False)
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
me = Base.classes.measurement
stat = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c['type'])
```

```python
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Find the most recent date in the data set.
session.query(me.date).order_by(me.date.desc()).all()
```

```python

# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
# Calculate the date one year from the last date in data set.
qry = session.query(me.date, me.prcp).filter(me.date >= "2016-08-23").filter(me.date <= "2017-08-23").all()
qry
# Perform a query to retrieve the data and precipitation scores


# Save the query results as a Pandas DataFrame and set the index to the date column


# Sort the dataframe by date


# Use Pandas Plotting with Matplotlib to plot the data


```

```python
byemon = pd.DataFrame(qry)
spymon = byemon.rename(columns={0: "date",1: "prcp"})
spymon
```

```python
paimore = spymon.sort_values(by=["date"])
pai = paimore.dropna()
pai
```

```python
paimon = pai.set_index("date")
paimon = paimon.fillna(0)
paimon
```

```python
plt.plot(paimon.index, paimon["prcp"], marker="o", label="precipitation", markersize=8, linewidth=1)
plt.legend(loc="upper right", prop={'size': 10})
plt.xticks(np.arange(0, 390, 39), rotation=90)
plt.xlabel("Date", font={'size': 14})
plt.ylabel("Precipitation", font={'size': 14})
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data 
data = pd.DataFrame({'mean' : [paimon["prcp"].mean()],
        'variance' : [paimon["prcp"].var()],
        'std' : [paimon["prcp"].std()],
        'count' : [paimon["prcp"].count()],
        'min' : [paimon["prcp"].min()],
        '25%' : [paimon["prcp"].quantile(0.25)],
        '50%' : [paimon["prcp"].median()],
        '75%' : [paimon["prcp"].quantile(0.75)],
        'max' : [paimon["prcp"].max()],
        'sem' : [paimon["prcp"].sem()]})
data
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
engine.execute("SELECT COUNT(station) FROM station;").all()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
engine.execute("SELECT station.station, station.name, count(measurement.station) FROM station, measurement WHERE station.station=measurement.station GROUP BY station.station, station.name ORDER BY count(measurement.station) desc;").all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
print(session.query(me.station, me.tobs).filter(me.station == 'USC00519281').order_by(me.tobs).limit(1).one())
print(session.query(me.station, me.tobs).filter(me.station == 'USC00519281').order_by(me.tobs.desc()).limit(1).one())
print(session.query(me.station, func.avg(me.tobs)).filter(me.station == 'USC00519281').one())
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
hist = session.query(me.station, me.tobs).filter(me.station == 'USC00519281').filter(me.date >= "2016-08-23").filter(me.date <= "2017-08-23").all()
hist
```

```python
query = pd.DataFrame(hist).rename(columns={0: "station",1: "tobs"})
query
```

```python
plt.hist(query["tobs"], bins=12, label="tobs")
plt.xlabel("Temperature", font={'size': 14})
plt.ylabel("Precipitation", font={'size': 14})
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
